```python
from PyPDF2 import PdfReader
from typing import List, Tuple
from io import BytesIO
import re
```

```python
def parse_pdf(file: BytesIO, filename: str) -> Tuple[List[str], str]:
    pdf = PdfReader(file)
    output = []
    for page in pdf.pages:
        text = page.extract_text()
        if text:  # Ensure text is not None
            # Fix broken words hyphenated across lines
            text = re.sub(r"(\w+)-\n(\w+)", r"\1\2", text)
            # Remove single newlines not followed by spaces
            text = re.sub(r"(?<!\n\s)\n(?!\s\n)", " ", text.strip())
            # Reduce multiple newlines to a single one
            text = re.sub(r"\n\s*\n", "\n\n", text)
            output.append(text)
    return output, filename

# Example usage
with open('Haor-Master-Plan-Volume-1.pdf', 'rb') as f:
    file = BytesIO(f.read())
    text, filename = parse_pdf(file, 'sample.pdf')
    for page_num, page_text in enumerate(text):
        print(f"Page {page_num + 1}:")
        print(page_text)
        print("\n" + "="*50 + "\n")
```
